## 1. Variables con significado

En un algoritmo de análisis de datos podemos describir cada variable desde
**dos perspectivas diferentes**:

### A. Según su origen

| Categoría | Función | Ejemplo |
|---|---|---|
| Dato original | Conserva información recibida | `ventas_diarias` |
| Constante de referencia | Representa un valor que no debe cambiar durante el análisis | `META_SEMANAL` |
| Variable derivada | Se calcula utilizando otros datos | `total_ventas` |

### B. Según su propósito

| Categoría | Función | Ejemplo |
|---|---|---|
| Indicador | Resume o interpreta los datos | `promedio_diario` |
| Variable lógica | Contesta una pregunta con `True` o `False` | `meta_cumplida` |
| Salida interpretada | Comunica una conclusión | `conclusion` |

Una variable puede pertenecer a una categoría de cada perspectiva. Por ejemplo,
`promedio_diario` es una **variable derivada por su origen** y un
**indicador por su propósito**.

Usaremos mayúsculas para constantes y `snake_case` para las demás variables.

````python
# Ejemplo guiado: variables de un análisis comercial
ventas_diarias = [125000, 98000, 147000, 132000, 110000]
META_SEMANAL = 600000

total_ventas = sum(ventas_diarias)
cantidad_dias = len(ventas_diarias)
promedio_diario = total_ventas / cantidad_dias
meta_cumplida = total_ventas >= META_SEMANAL

print(f"Total de ventas: ₡{total_ventas:,.2f}")
print(f"Promedio diario: ₡{promedio_diario:,.2f}")
print(f"¿Se cumplió la meta?: {meta_cumplida}")
````

## 2. Dependencias y trazabilidad

Una variable derivada depende de otras variables. El orden del algoritmo es
importante porque Python necesita conocer los datos antes de calcular un
resultado.

```text

       |                              |
       +----> cantidad_dias ----------+----> promedio_diario
       |
       +----> total_ventas ----+
                               +----> meta_cumplida
META_SEMANAL ------------------+
```

El promedio depende de `total_ventas` y `cantidad_dias`. La decisión depende de
`total_ventas` y `META_SEMANAL`.

Si cambia una venta, deben recalcularse todos los indicadores que dependen de
ella.

````python
##semana 5 18062026
#Gustavo Bellido O.
#Promedio de Ventas
#

ventas_semana = [145000, 118000, 162000, 137000, 128000]
META_SEMANAL = 700000


cantidad_dias = len(ventas_semana)
total_ventas = sum(ventas_semana)
promedio_ventas = total_ventas / cantidad_dias
venta_mayor = max(ventas_semana)
venta_menor = min(ventas_semana)
rango_ventas = venta_mayor - venta_menor


porcentaje_meta = (total_ventas / META_SEMANAL) * 100


if total_ventas >= META_SEMANAL:
    diferencia = total_ventas - META_SEMANAL
    mensaje = f"La empresa cumplió la meta semanal y superó la meta en ₡{diferencia:,}."
else:
    diferencia = META_SEMANAL - total_ventas
    mensaje = f"La empresa no cumplió la Meta Semanal hizo falta ₡{diferencia:,}."


print("===== REPORTE DE VENTAS SEMANALES =====")

print(f"Ventas registradas: {ventas_semana}")
print(f"Cantidad de días: {cantidad_dias}")
print(f"Total vendido: ₡ {total_ventas:,}")
print(f"Promedio diario: ₡ {promedio_ventas:,.2f}")
print(f"Venta mayor: ₡ {venta_mayor:,}")
print(f"Venta menor: ₡ {venta_menor:,}")
print(f"Rango de ventas: ₡{rango_ventas:,}")

print("\nMETA SEMANAL")
print(f"Esta es la Meta establecida por la empresa: ₡ {META_SEMANAL:,}")
print(f"Porcentaje de la meta alcanzado : {porcentaje_meta:.2f}%")

print("\nCONCLUSIÓN")
print(mensaje)
#
#
print ("Tipo de variable")
print(f"El tipo de variable de ventas_semana es: {type(ventas_semana)}")
print(f"El tipo de variable de META_SEMANAL es: {type(META_SEMANAL)}")
print(f"El tipo de variable de cantidad_dias es: {type(cantidad_dias)}")
print(f"El tipo de variable de total_ventas es: {type(total_ventas)}")
print(f"El tipo de variable de promedio_ventas es: {type(promedio_ventas)}")
print(f"El tipo de variable de venta_mayor es: {type(venta_mayor)}")
print(f"El tipo de variable de venta_menor es: {type(venta_menor)}")
print(f"El tipo de variable de rango_ventas es: {type(rango_ventas)}")
print(f"El tipo de variable de porcentaje_meta es: {type(porcentaje_meta)}")
````
