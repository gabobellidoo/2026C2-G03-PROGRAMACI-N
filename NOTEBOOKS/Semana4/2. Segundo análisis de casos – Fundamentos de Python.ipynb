# Problema #4: Validador de compra con descuento

## Problema

# Se solicita un usuario, un monto y se pregunta si es cliente frecuente, respuesta es tru recibe descuento, respuesta es false no recibe descuento

## Entrada

- total= float (0)

cliente = input("Ingrese el ID del cliente: ")
monto = float(input("Ingrese el monto de la compra: $"))
cliente_frecuente = input("¿Es cliente frecuente? (S/N): ")

## Proceso

- if cliente_frecuente.upper() == "S":
    descuento = monto * 0.10
    total = monto - descuento
    print("\nSe aplica un descuento del 10%.")
else:
    descuento = 0
    total = monto
    print("\nNo se aplica descuento.")

## Salida esperada

- print("\n----- FACTURA -----")
print(f"Cliente: {cliente}")
print(f"Monto de la compra: ${monto:.2f}")
print(f"Descuento: ${descuento:.2f}")
print(f"Total a pagar: ${total:.2f}")

## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|-----------|------------------|--------------------|--------------------|
| Caso normal | | | |
| Caso límite | | | |
| Caso inválido | | | |


```python
####semana 4 11062026
#Gustavo Bellido O.
#Cliente frecuente
#

try:
    total= float (0)

    cliente = int(input("Ingrese el ID del cliente: ").strip())
    monto = float(input("Ingrese el monto de la compra: $").strip())
    cliente_frecuente = input("¿Es cliente frecuente? (S/N): ").strip().upper()

    if cliente_frecuente not in ("S", "N"):
        raise ValueError("Respuesta inválida")

    if cliente_frecuente.upper() == "S":
     
        descuento = monto * 0.10
        total = monto - descuento
        print("\nSe aplica un descuento del 10%.")
    else:
        descuento = 0
        total = monto
        print("\nNo se aplica descuento.")

    print("\n----- FACTURA -----")
    print(f"Cliente: {cliente}")
    print(f"Monto de la compra: ${monto:.2f}")
    print(f"Descuento: ${descuento:.2f}")
    print(f"Total a pagar: ${total:.2f}")

except ValueError:
   print("Error: Debe ingresar únicamente valores que se le solicitan en los campos.")
```
