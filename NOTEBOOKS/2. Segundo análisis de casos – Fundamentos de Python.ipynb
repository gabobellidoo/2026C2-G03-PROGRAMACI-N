## Problema #1: Calculadora de propina inteligente

## Problema

Se debe calcular el total de una factura a partir del monto del consumo y el porcentaje de propina que el cliente desea dejar.

## Entrada

- Monto del consumo: `float`
- Porcentaje de la propina: `float`

## Proceso

- Validar que el monto y el porcentaje sean valores numéricos y no negativos.
- Calcular la propina: `propina = consumo * porcentaje / 100`.
- Sumar el consumo más la propina para obtener el total.

## Salida esperada

- Mostrar el monto del consumo, la propina calculada y el total a pagar en formato de factura.

## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|-----------|------------------|--------------------|--------------------|
| Caso normal | 100, 15 | Propina $15.00, total $115.00 | |
| Caso límite | 0, 10 | Propina $0.00, total $0.00 | |
| Caso inválido | "abc", 15 | Mensaje de error de entrada | |

```python
# semana 4 11062026
# Gustavo Bellido O.
# Calculadora de propina inteligente

try:
    total_consumo = float(input("Ingrese el monto del consumo: $"))
    porcentaje_propina = float(input("Ingrese el porcentaje de la propina: % "))

    if total_consumo < 0 or porcentaje_propina < 0:
        raise ValueError("Los valores no pueden ser negativos.")

    propina = total_consumo * (porcentaje_propina / 100)
    total_pagar = total_consumo + propina

    print("\n----- DETALLE DE LA FACTURA -----")
    print(f"Monto del consumo:     $ {total_consumo:.2f}")
    print(f"Propina ({porcentaje_propina:.0f}%):        $ {propina:.2f}")
    print(f"Total a cancelar:      $ {total_pagar:.2f}")
except ValueError:
    print("ERROR: Ingrese un monto y un porcentaje válidos.")
```

## Problema #2: Detector de número par/impar

## Problema

Determinar si un número entero ingresado por el usuario es par o impar.

## Entrada

- Un número entero ingresado por teclado.

## Proceso

- Validar que la entrada sea un entero.
- Usar el residuo de la división por 2: `numero % 2`.
- Si el residuo es 0, el número es par; de lo contrario, es impar.

## Salida esperada

- Mensaje indicando si el número es par o impar.

## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|-----------|------------------|--------------------|--------------------|
| Caso normal | 7 | "7 es Impar" | |
| Caso límite | 0 | "0 es Par" | |
| Caso inválido | "cinco" | Mensaje de error de entrada | |

```python
# semana 4 11062026
# Gustavo Bellido O.
# Detector de número par/impar

try:
    numero = int(input("Ingrese un número entero: "))
    tipo = "Par" if numero % 2 == 0 else "Impar"
    print(f"El número ingresado {numero} es {tipo}.")
except ValueError:
    print("ERROR: Debe ingresar un número entero válido.")
```

# Problema #3: Clasificador de edad

## Problema

Clasificar a una persona según su edad en: menor, adulto o adulto mayor.

## Entrada

- Edad ingresada por el usuario como número entero.

## Proceso

- Validar que la edad sea un número entero no negativo.
- Si la edad es menor o igual a 17, clasificar como "Menor".
- Si la edad está entre 18 y 59, clasificar como "Adulto".
- Si la edad es 60 o más, clasificar como "Adulto mayor".

## Salida esperada

- Mensaje con la categoría de edad correspondiente.

## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|-----------|------------------|--------------------|--------------------|
| Caso normal | 25 | "Adulto" | |
| Caso límite | 60 | "Adulto mayor" | |
| Caso inválido | -5 | Mensaje de edad inválida | |

```python
try:
    edad = int(input("Ingrese su edad: "))

    if edad < 0:
        raise ValueError("Edad negativa")

    if edad <= 17:
        categoria = "Menor"
    elif edad <= 59:
        categoria = "Adulto"
    else:
        categoria = "Adulto mayor"

    print(f"Con {edad} años, la persona es: {categoria}.")
except ValueError:
    print("ERROR: Debe ingresar una edad válida (entero no negativo).")
```

# Problema #4: Validador de compra con descuento

## Problema

Calcular el descuento aplicado a una compra según si el cliente es frecuente y el monto de la compra.

## Entrada

- Nombre del cliente.
- Monto de la compra (`float`).
- Si es cliente frecuente (`S`/`N`).

## Proceso

- Validar el monto como número no negativo.
- Normalizar la respuesta de cliente frecuente a mayúsculas.
- Aplicar descuento según las reglas:
  - Cliente frecuente y monto >= 100: 15%
  - Cliente frecuente y monto < 100: 10%
  - No frecuente y monto >= 200: 5%
  - De lo contrario: 0%
- Calcular el total con descuento.

## Salida esperada

- Mostrar cliente, monto original, descuento y total a pagar.

## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|-----------|------------------|--------------------|--------------------|
| Caso normal | Cliente frecuente, 150 | Descuento 15%, total 127.50 | |
| Caso límite | No frecuente, 200 | Descuento 5%, total 190.00 | |
| Caso inválido | "abc" en monto | Mensaje de error de entrada | |

```python
cliente = input("Ingrese el nombre del cliente: ").strip()
frecuente = input("¿Es cliente frecuente? (S/N): ").strip().upper()

try:
    monto = float(input("Ingrese el monto de la compra: $"))
    if monto < 0:
        raise ValueError("Monto negativo")

    if frecuente == "S":
        descuento = 0.15 if monto >= 100 else 0.10
    elif frecuente == "N":
        descuento = 0.05 if monto >= 200 else 0.0
    else:
        raise ValueError("Respuesta de cliente frecuente inválida")

    monto_descuento = monto * descuento
    total = monto - monto_descuento

    print("\n----- VALIDACIÓN DE LA COMPRA -----")
    print(f"Cliente: {cliente}")
    print(f"Monto original: $ {monto:.2f}")
    print(f"Descuento aplicado: {descuento * 100:.0f}%")
    print(f"Total a pagar: $ {total:.2f}")
except ValueError:
    print("ERROR: Ingrese un monto válido y una respuesta correcta (S/N).")
```
